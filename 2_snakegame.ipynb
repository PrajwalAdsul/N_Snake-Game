```python
import pygame
import random

pygame.init()
window_width = 1300
window_height = 700
win = pygame.display.set_mode((window_width, window_height))
pygame.display.set_caption("N way Snake Game")

snake1x = 50
snake1y = 50

snake2x = 100
snake2y = 200
x = 50
y = 50
width1 = 10
width2 = 10
height = 10
vel = 25

snake1w = 10
snake1h = 10
snake2w = 10
snake2h = 10

x1 = 100
y1 = 100

x2 = 200
y2 = 200

x2_list = [50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650]
y2_list = x2_list

direction1 = 'r'
direction2 = 'd'
white = (255, 255, 255) 
green = (0, 255, 0) 
blue = (0, 0, 128) 
  

count1 = 0
count2 = 0
c = count1 + count2

font = pygame.font.Font('freesansbold.ttf', 32)
text = font.render(str(c), True, green, blue) 
textRect = text.get_rect() 
textRect.center = (200, 50)
quit = False

run = True
while run:
    if(quit):
        text = font.render("game over", True, green, blue)
        win.blit(text, textRect)
        pygame.display.update()
    else:
        text = font.render(str(c), True, green, blue)
        win.blit(text, textRect)
        pygame.display.update()

        pygame.time.delay(100)

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                run = False

        keys = pygame.key.get_pressed()

        if keys[pygame.K_LEFT]:
            snake1x = snake1x - vel
            direction1 = 'l'
        elif keys[pygame.K_RIGHT]:
            snake1x = snake1x + vel
            direction1 = 'r'
        elif keys[pygame.K_UP]:
            snake1y = snake1y - vel;
            direction1 = 'u'
        elif keys[pygame.K_DOWN]:
            snake1y = snake1y + vel
            direction1 = 'd'
        elif direction1 == 'l':
            snake1x = snake1x - vel
        elif direction1 == 'r':
            snake1x = snake1x + vel
        elif direction1 == 'u':
            snake1y = snake1y - vel
        else:
            snake1y = snake1y + vel


        win.fill((0,0,0))


        keys1 = pygame.key.get_pressed()

        if keys1[pygame.K_a]:
            snake2x = snake2x - vel 
            direction2 = 'l'
        if keys1[pygame.K_d]:
            snake2x = snake2x + vel
            direction2 = 'r'
        if keys1[pygame.K_w]:
            snake2y = snake2y - vel;
            direction2 = 'u'
        if keys1[pygame.K_s]:
            snake2y = snake2y + vel
            direction2 = 'd'
        elif direction2 == 'l':
            snake2x = snake2x - vel
        elif direction2 == 'r':
            snake2x = snake2x + vel
        elif direction2 == 'u':
            snake2y = snake2y - vel
        else:
            snake2y = snake2y + vel

        win.fill((0,0,0))

        #arguments win, color, dimensions
        pygame.draw.rect(win, (255, 0, 0), (snake1x, snake1y, snake1w, snake1h))

        pygame.draw.rect(win, (0, 255, 0), (snake2x, snake2y, snake2w, snake2h))
        
        if((abs(snake1x - x2) < 10 and abs(snake1y - y2) < 10)):
        #if((x == x2 and y == y2) or (x1 == x2 and y1 == y2)):
            x2 = random.choice(x2_list)
            y2 = random.choice(y2_list)
            count1 += 1

        if( (abs(snake2x - x2) < 20 and abs(snake2y - y2) < 20)):
            x2 = random.choice(x2_list)
            y2 = random.choice(y2_list)
            count2 += 1

        if(snake1x <= 0 or snake2x <= 0 or snake1y <= 0 or snake2y <= 0 or snake1x >= 1300 or snake1y >= 700 or snake2x >= 1300 or snake2y >= 700):
            text = font.render("your score is " + str(c), True, green, blue)
            win.blit(text, textRect)
            pygame.display.update()
            quit = True
            pygame.time.delay(1000)
            pygame.quit()

        pygame.draw.circle(win, (0, 0, 255), (x2, y2), 10)
        c = count1 + count2

        pygame.display.update()

pygame.quit()
```
